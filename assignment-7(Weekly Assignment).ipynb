```python
len(odata[odata.Medal=='Gold'])
```

# 11. Display the male gold medal winners for the 100m Track?

```python
male_gold_medal_winners=odata[(odata.Gender=='Men')&(odata.Event=='100m')&(odata.Medal=='Gold')][['Athlete','Event','Medal']]
male_gold_medal_winners
```

# 12.Which three countries have won the most medals in recent years (from 1984 to 2008)?

```python
odata[(odata.Edition>=1984)&(odata.Edition<=2008)]['NOC'].value_counts().head(3)
```

# 13. Which country has won the most men's gold medals in singles badminton over the years? Sort the results alphabetically by the player's names.


```python
odata[(odata.Medal=='Gold')&(odata.Gender=='Men')&(odata.Event=='singles') & (odata.Discipline=='Badminton')].sort_values(['Athlete'])
```

# 14. Display the women gold medal winners for the 100m Track?

```python
women_gold_medal_winners=odata[(odata.Gender=='Women')&(odata.Medal=='Gold')&(odata.Event=='100m')][['Athlete','Gender','Event','Medal']]
women_gold_medal_winners
```

# 15.list out the Sliver medal winner @ year 2008?

```python
silver_2008_winner=odata[(odata.Medal=='Silver')&(odata.Edition==2008)][['Athlete','Edition','Medal']]
silver_2008_winner
```

# 16.Group by Countries and apply aggregate function min,max,count on over the year

```python
odata.groupby('NOC')['Edition'].agg([min,max,'count'])
```
